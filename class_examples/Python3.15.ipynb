````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

````python
df = pd.read_csv('sensor_data.csv')
df.head()
````

````python
df.info()
````

````python
#build the temperature vs humidity by location plot

dfA = df[df.Location == 'Location_A']
dfB = df[df.Location == 'Location_B']

fig, ax = plt.subplots(2,1, sharex = True, sharey = True)

ax[0].scatter(dfA.Temperature_C, dfA.Humidity, marker = '*', c = 'm')
ax[1].scatter(dfB.Temperature_C, dfB.Humidity, marker = '+', c = 'b')
#
````
